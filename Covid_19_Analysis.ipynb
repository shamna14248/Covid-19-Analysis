## **Import Libraries and Load Data**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## **Load and Prepare COVID-19 data**

```python
dataset=pd.read_csv("/content/covid19_Confirmed_dataset.csv")
```

```python
dataset.head()
```

```python
dataset.shape
```

## **Clean and Aggregate COVID-19 Data**

```python
df=dataset.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
corona_dataset_aggregated=dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

## **Analyze COVID-19 Trends by Country**

```python

corona_dataset_aggregated.loc["China"].apply(pd.to_numeric, errors='coerce').plot()
corona_dataset_aggregated.loc["India"].apply(pd.to_numeric, errors='coerce').plot()
corona_dataset_aggregated.loc["Spain"].apply(pd.to_numeric, errors='coerce').plot()
plt.legend()
```

```python
corona_dataset_aggregated.loc["India"].apply(pd.to_numeric, errors='coerce').diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].apply(pd.to_numeric, errors='coerce').diff().max()
```

```python
countries=list(corona_dataset_aggregated.index)
max_infection_rates=[]
corona_dataset_aggregated = corona_dataset_aggregated.apply(pd.to_numeric, errors='coerce')
for c in countries:
  max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["Max_infection_rates"]=max_infection_rates
```

```python
corona_dataset_aggregated
```

## **Calculate Maximum Infection Rates**

```python
corona_data=pd.DataFrame(corona_dataset_aggregated["Max_infection_rates"])
```

```python
corona_data
```

## **Load and Prepare Happiness Report Data**

```python
happiness_report=pd.read_csv("worldwide_happiness_report.csv")
```

```python
happiness_report
```

```python
useless_cols=["Overall rank","Score","Generosity","Perceptions of corruption"]
happiness_report.drop(useless_cols,axis=1,inplace=True)
happiness_report.head()
```

```python
happiness_report.set_index("Country or region",inplace=True)
happiness_report.head()
```

```python
corona_data.shape
```

```python
happiness_report.shape
```

## **Merge COVID-19 and Happiness Data**

```python
data = corona_data.join(happiness_report, how="inner")
data
```

## **Correlation Analysis**

```python
data.corr()
```

## **Visualize Relationships Between Happiness Factors and Infection Rates**

```python


# Assuming 'data' is your DataFrame containing the columns 'GDP per capita' and 'Max_infection_rates'
x = data["GDP per capita"]
y = np.log(data["Max_infection_rates"])  # Applying logarithm to y

# Corrected scatterplot call with keyword arguments
sns.scatterplot(x=x, y=y)
```

```python
# Regression plot with keyword arguments
sns.regplot(x=x, y=y)
```
